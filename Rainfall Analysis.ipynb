```python
df_far = pd.read_excel("Faridpur_rainfall.xls")

# df.head(20)

year_range = np.arange(1948,1988,1)

filt = ~df_far["YEAR"].isin(year_range)

df_far = df_far[filt]

df_far = df_far.sort_values(by = ["YEAR","MONTH"])

df_far = df_far.reset_index(drop = True)

# df

new_df_far = df_far.drop(df_far.columns[4:],axis = 1)

monthly_max_far = new_df_far.groupby("MONTH").max()

monthly_avg_far = new_df_far.groupby("MONTH").mean()

monthly_max_far["Average"] = monthly_avg_far["SUMOFTSVALUE_UPDATED"]

monthly_max_far.rename(columns = {"SUMOFTSVALUE_UPDATED":"Max"}, inplace = True)

# monthly_max.to_excel("monthly_analysis_tangail.xls")



# monthly_max

monthly_max_far.reset_index()

import calendar

monthly_max_far['MONTH'] = df_far['MONTH'].apply(lambda x: calendar.month_abbr[x-1])

monthly_max_far.drop(["STATIONID","YEAR"], axis = 1, inplace = True)

monthly_max_far.to_excel("monthly_faridpur.xls")

yearly_sum_far = new_df_far.groupby('YEAR').sum()

# yearly_sum

yearly_sum_far.drop(["STATIONID","MONTH"],axis = 1, inplace = True)

yearly_sum_far.to_excel("faridpur_yearly_analysis.xls")
```
