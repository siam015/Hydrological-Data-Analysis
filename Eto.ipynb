```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel("faridpur_eto.xls")
```

```python
df
```

```python
df_max = df.loc[:,["YEAR","MONTH","MAXOFETO"]]
df_avg = df.loc[:,["YEAR","MONTH","AVGOFETO"]]
```

```python
monthly_max = df_max.groupby("MONTH").mean()
monthly_avg = df_avg.groupby("MONTH").mean()
```

```python
monthly_max.reset_index(inplace = True)
```

```python
monthly_analysis = monthly_max.merge(monthly_avg, on = "MONTH", how = "inner")
```

```python
monthly_analysis.drop(["YEAR_x","YEAR_y"],axis = 1, inplace = True)
```

```python
import calendar
monthly_analysis['MONTH'] = monthly_analysis['MONTH'].apply(lambda x: calendar.month_abbr[x])
```

```python
monthly_analysis.to_excel("faridpur_eto_monthly_analysis.xls")
```
