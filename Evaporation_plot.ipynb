```python
import seaborn as sns
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
```

```python
monthly_evap = pd.read_excel("faridpur_monthly_evaporation_analysis.xls")
```

```python
monthly_evap
```

```python
monthly_evap = monthly_evap.drop(monthly_evap.columns[monthly_evap.columns.str.contains('unnamed',case = False)],axis = 1)
```

```python
monthly_evap.columns = ["Month","Maximum","Average"]
```

```python
month_order = ["Apr", "May","Jun","Jul","Aug","Sep","Oct","Nov","Dec","Jan","Feb","Mar"]

# sns.set(rc={'figure.figsize':(5,3.5)})
# sns.set_style("whitegrid")

# fig, ax = plt.subplots()
# ax = sns.barplot(x="Month", hue = None, y="Maximum",data=monthly_evap, order=["Apr", "May","Jun","Jul","Aug","Sep","Oct","Nov","Dec","Jan","Feb","Mar"], color = "royalblue", label = "Maximum")

new_df = monthly_evap.set_index(monthly_evap["Month"]).loc[month_order]

plt.plot(new_df["Average"], color = 'salmon',marker = 'o', label = "Average")

plt.legend()

plt.yticks(np.arange(0,11,1))

plt.xlabel("Month", fontweight="semibold")
plt.ylabel("Evaporation (mm/day)", fontweight="semibold")
plt.tight_layout()
plt.savefig("faridpur_evaporation.jpg",dpi = 500)
plt.show()
```
