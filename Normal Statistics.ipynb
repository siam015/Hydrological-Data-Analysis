```python
import pandas as pd
import numpy as np

import sklearn.metrics as metrics
```

```python
df = pd.read_csv("Soil_Moisture_Dataset_Final.csv",index_col = 0)
```

```python
df.info()
```

```python
Observed_sm = df["5cm"]
```

```python
SMAP_sm = df["7_day_avg_sm"]
```

```python
rmse = metrics.mean_squared_error(Observed_sm, SMAP_sm, squared = False)
 
r2 = metrics.r2_score(Observed_sm, SMAP_sm)
```

```python
print("RMSE:", rmse)
print("R-Squared:", r2)
```
