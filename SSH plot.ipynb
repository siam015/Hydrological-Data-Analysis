```python
import seaborn as sns
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
```

```python
monthly_ssh = pd.read_excel("monthly_ssh_analysis_faridpur.xls")
```

```python
monthly_ssh
```

```python
monthly_ssh = monthly_ssh.drop(monthly_ssh.columns[monthly_ssh.columns.str.contains('unnamed',case = False)],axis = 1)
```

```python
monthly_ssh.columns = ["Month","Average"]
```

```python
sns.set(rc={'figure.figsize':(5,3.5)})
sns.set_style("whitegrid")

ax = sns.barplot(x="Month", hue = None, y="Average",data=monthly_ssh, order=["Apr", "May","Jun","Jul","Aug","Sep","Oct","Nov","Dec","Jan","Feb","Mar"], color = "mediumslateblue", label = "Average", )
plt.legend()

plt.yticks(np.arange(0,10,1))

plt.grid(visible = True, axis = 'y',  alpha = 0.6)
plt.xlabel("Month", fontweight="semibold")
plt.ylabel("Sunshine (hour/day)", fontweight="semibold")

plt.tight_layout()
plt.savefig("faridpur_ssh.jpg",dpi = 500)
plt.show()
```
