```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
df = pd.read_excel("Faridpur_temp.xls")
```

```python
df.info()
```

```python
df_new  = df.loc[:,["STATIONID","YEAR","MONTH","MAXOFMAXTEMP_UPDATED","MINOFMAXTEMP_UPDATED","MAXOFMINTEMP_UPDATED","MINOFMINTEMP_UPDATED","AVGOFMAXTEMP_UPDATED","AVGOFMINTEMP_UPDATED"]]
```

```python
df_new
```

```python
filt = df_new["YEAR"] != 1987
df_new = df_new[filt]
```

```python
df_new
```

```python
df_new = df_new.sort_values(by = ["YEAR","MONTH"])
```

```python
df_new
```

```python
monthly_max = df_new.groupby("MONTH").max()
```

```python
monthly_max
```

```python
monthly_min = df_new.groupby("MONTH").min()
```

```python
monthly_avg = df_new.groupby("MONTH").mean()
```

```python
monthly_analysis = monthly_max.loc[:,["STATIONID","MAXOFMAXTEMP_UPDATED","MAXOFMINTEMP_UPDATED"]]
```

```python
monthly_analysis["MINOFMAXTEMP_UPDATED"] = monthly_min.loc[:,["MINOFMAXTEMP_UPDATED"]]
monthly_analysis["MINOFMINTEMP_UPDATED"] = monthly_min.loc[:,["MINOFMINTEMP_UPDATED"]]
```

```python
monthly_analysis["avg_max"] = monthly_avg.loc[:,["AVGOFMAXTEMP_UPDATED"]]
monthly_analysis["avg_min"] = monthly_avg.loc[:,["AVGOFMINTEMP_UPDATED"]]
```

```python
monthly_analysis
```

```python
import calendar
monthly_analysis = monthly_analysis.reset_index()
monthly_analysis['MONTH'] = monthly_analysis['MONTH'].apply(lambda x: calendar.month_abbr[x])
```
