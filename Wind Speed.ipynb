```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel("faridpur_ws.xls")
```

```python
df.info()
```

```python
df_new = df.loc[:,["STATIONID","YEAR","MONTH","MAXOFTSVALUE_UPDATED"]]
```

```python
df_new["MAXOFTSVALUE_UPDATED"] = df_new["MAXOFTSVALUE_UPDATED"]/24
```

```python
df_new = df_new.sort_values(by = ["YEAR","MONTH"])
```

```python
df_new
```

```python
filt = df_new["YEAR"] != 1987
df_new = df_new[filt]
```

```python
monthly_avg = df_new.groupby("MONTH").mean()
monthly_max = df_new.groupby("MONTH").max()
```

```python
monthly_avg = monthly_avg.reset_index()
```

```python
monthly_avg.info()
```

```python
monthly_analysis = monthly_avg.loc[:,["MONTH","MAXOFTSVALUE_UPDATED"]]
```

```python
monthly_analysis.rename({"MAXOFTSVALUE_UPDATED": "Max_avg"}, inplace = True)
```

```python
monthly_analysis_merged = monthly_analysis.merge(monthly_max,on = "MONTH")
```

```python
monthly_analysis_merged.info()
```

```python
monthly_analysis_merged.rename({"MAXOFTSVALUE_UPDATED_x":"Max_avg","MAXOFTSVALUE_UPDATED_y":"max"})
```

```python
monthly_analysis_merged.drop(["STATIONID","YEAR"], axis = 1, inplace = True)
```

```python
import calendar
monthly_analysis_merged['MONTH'] = monthly_analysis_merged['MONTH'].apply(lambda x: calendar.month_abbr[x])
```

```python
monthly_analysis_merged
```

```python
monthly_analysis_merged.info()
```

```python
monthly_analysis_merged.to_excel("faridpur_monthly_analysis.xls")
```
