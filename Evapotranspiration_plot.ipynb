```python
import seaborn as sns
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
```

```python
monthly_eto = pd.read_excel("faridpur_eto_monthly_analysis.xls")
```

```python
monthly_eto
```

```python
monthly_eto = monthly_eto.drop(monthly_eto.columns[monthly_eto.columns.str.contains('unnamed',case = False)],axis = 1)
```

```python
monthly_eto.info()
```

```python
monthly_eto.drop(["AVGOFETO"], axis = 1, inplace = True)
```

```python
monthly_eto.columns = ["Month","Average"]
```

```python
sns.set(rc={'figure.figsize':(5,3.5)})
sns.set_style("whitegrid")

fig, ax = plt.subplots()
ax = sns.barplot(x="Month", hue = None, y="Average",data=monthly_eto, order=["Apr", "May","Jun","Jul","Aug","Sep","Oct","Nov","Dec","Jan","Feb","Mar"], color = "darkcyan", label = "Average")

plt.legend()

plt.yticks(np.arange(0,7,1))
plt.grid(visible = True, axis = 'y',  alpha = 0.9)

plt.xlabel("Month", fontweight="semibold")
plt.ylabel("evapotranspiration (mm/day)", fontweight="semibold")
plt.tight_layout()
plt.savefig("faridpur_eto.jpg",dpi = 500)
plt.show()
```
