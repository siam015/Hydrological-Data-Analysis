```python
monthly  = pd.read_excel("monthly_faridpur.xls")

yearly = pd.read_excel("faridpur_yearly_analysis (1).xls")
```

```python
monthly = monthly.drop(["MONTH"],axis = 1)
```

```python
tidy = monthly.melt(id_vars='MONTH.1').rename(columns=str.title)
```

```python
tidy
```

```python
labels = ["Maximum","Average"]
```

```python
sns.set(rc={'figure.figsize':(5,3)})
sns.set_style("whitegrid")
ax_new = sns.barplot(x="Month.1", y="Value", hue = 'Variable' ,data=tidy, order=["Apr", "May","Jun","Jul","Aug","Sep","Oct","Nov","Dec","Jan","Feb","Mar"], palette="Set2")

plt.yticks([0,100,200,300,400,500,600,700,800,900,1000])
plt.grid(visible = True, axis = 'y',  alpha = 0.6)

ax_new.legend()

plt.xlabel("Month", fontweight="semibold")
plt.ylabel("Rainfall (mm)", fontweight="semibold")
plt.tight_layout()
plt.savefig("faridpur_monthly_rainfall.jpg", dpi = 500)
plt.show()
```
