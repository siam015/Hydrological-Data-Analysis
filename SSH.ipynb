```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel("faridpur_ssh.xls")
```

```python
filt = df["YEAR"] != 1987
df_new = df[filt]
```

```python
df_new.info()
```

```python
monthly_avg = df_new.groupby("MONTH").mean()
```

```python
monthly_avg = monthly_avg.reset_index()
```

```python
import calendar
monthly_avg['MONTH'] = monthly_avg['MONTH'].apply(lambda x: calendar.month_abbr[x])
```

```python
monthly_avg.drop(["STATIONID","YEAR"],axis = 1, inplace = True)
```

```python
monthly_avg.to_excel("monthly_ssh_analysis_faridpur.xls")
```
